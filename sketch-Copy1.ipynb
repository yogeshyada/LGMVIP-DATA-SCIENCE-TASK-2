# Image to Pencil Sketch with Python

```python
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read the image

```python
image = cv2.imread('C:/Yogesh/Wallpapers/mustang.jpg')
```

```python
cv2.imshow("Car", image)
cv2.waitKey(0)
```

# Plotting the Original Image

```python
img = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.show()
```

# Converting the Original Image to GrayScale

```python
gray_img = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(gray_img, cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.show()
```

```python
invert_img = cv2.bitwise_not(gray_img)
plt.figure(figsize=(8,8))
plt.imshow(invert_img, cmap="gray")
plt.axis("off")
plt.title("Imverted Image")
plt.show()
```

# Blurring the Image to make the sketch smooth

```python
smoothing_img = cv2.GaussianBlur(invert_img ,(21,21),sigmaX=0,sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(smoothing_img, cmap="gray")
plt.axis("off")
plt.title("Smoothing Image")
plt.show()
```

# Converting the image to Pencil Sketch

```python
sketch_img = cv2.divide(gray_img ,255 - smoothing_img, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(sketch_img, cmap="gray")
plt.axis("off")
plt.title("Sketch Image")
plt.show()
```

# Conversion in one frame

```python
plt.figure(figsize=(20,15))
plt.subplot(1,5,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(gray_img, cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(invert_img, cmap="gray")
plt.axis("off")
plt.title("Imverted Image")
plt.subplot(1,5,4)
plt.imshow(smoothing_img, cmap="gray")
plt.axis("off")
plt.title("Smoothing Image")
plt.subplot(1,5,5)
plt.imshow(sketch_img, cmap="gray")
plt.axis("off")
plt.title("Sketch Image")
plt.show()
```
